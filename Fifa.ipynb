```python
plt.figure(figsize=(8,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color="b")
plt.show() 
```

*From this bar chart we can see that most nationality count is from England*

```python
player_salary = fifa[['short_name','wage_eur']]
player_salary.head()
```

```python
player_salary=player_salary.sort_values(by=['wage_eur'],ascending=False)
player_salary.head()
```

```python
plt.figure(figsize=(15,6))
plt.bar(list(player_salary['short_name'])[0:6],list(player_salary['wage_eur'])[0:6],color="g")
plt.show()
```
